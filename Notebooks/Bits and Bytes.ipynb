## Playing with bits and bytes

#### Sometimes one has to go deep and deal with the stored bits and bytes rather than variables. Here, I have some functions that might be useful in such cases.

### Bin: returns the integer in binary format

```python
a = 30
a_bin = bin(a)

print("a in binary: {}".format(a_bin))
print("Type: {}".format(type(a_bin)))
```

### `>>`: shifts the bits to the right


```python
b = a >> 1

b_bin = bin(b)

print("a: {} and b: {}".format(a, b))
print("a in binary: {}".format(a_bin))
print("b in binary: {}".format(b_bin))
```

### `<<`: shifts the bits to the left



```python
c = a << 1

c_bin = bin(c)

print("a: {} and c: {}".format(a, c))
print("a in binary: {}".format(a_bin))
print("c in binary: {}".format(c_bin))
```
